# Compute & Development Environment (Databricks)

## 1. Compute Clusters
- Group of virtual machines for data processing
- Powered by **Apache Spark**
- Two main nodes:
  - **Driver node** – manages execution
  - **Worker nodes** – process data in parallel (distributed processing)
- Cluster configuration:
  - Databricks Runtime
  - Node type (CPU / RAM)
  - Number of workers
  - Autoscaling

### Cluster Types
- **All-purpose cluster**
  - Development and notebooks
- **Job cluster**
  - Production jobs and pipelines
  - Created automatically and terminated after execution
- **SQL Warehouse**
  - Optimized for SQL analytics and BI tools

### Cluster Management
- **Autoscaling**
  - Automatically adjusts number of workers
- **Cluster policies**
  - Admin rules controlling cluster configuration

## 2. Notebooks

## Summary

- **Compute clusters** → provide computing power  
- **Notebooks** → environment for development and analysis  
- **Git integration** → version control and collaboration

A compute cluster is a group of virtual machines that process data using Apache Spark. The cluster is used to run code from notebooks, SQL queries, and data pipelines.

A cluster has two main types of nodes: driver node and worker nodes.

The driver node manages the execution of the program. It creates the execution plan and sends tasks to the worker nodes.

The worker nodes perform the actual data processing. They process different parts of the data in parallel. This approach is called distributed processing, and it allows the system to work with large datasets efficiently.

When creating a cluster, we must configure several parameters.
One important parameter is the Databricks Runtime, which includes Apache Spark and additional libraries.

Another parameter is the node type, which defines the CPU and memory of the machines.

We also define the number of worker nodes and the autoscaling settings.

Autoscaling allows the cluster to automatically increase or decrease the number of worker nodes depending on the workload. This helps optimize resource usage and reduce costs.

Databricks also supports cluster policies. These are rules created by administrators to control cluster configurations. For example, policies can limit the number of workers or restrict expensive machine types.

There are also different types of clusters.

All-purpose clusters are used for development and interactive work with notebooks.

Job clusters are used for automated workflows and production jobs. They are created when the job starts and removed after it finishes.

For SQL analytics and BI tools, Databricks provides SQL Warehouse, which is optimized for running SQL queries and dashboards.

Now I will explain notebooks.

A notebook is an interactive development environment where we can write code, run queries, and analyze data.

A notebook is organized into cells.

There are two main types of cells: code cells and markdown cells.

Code cells contain executable code. Markdown cells contain text and documentation.

Databricks notebooks support multiple programming languages, including Python, SQL, Scala, and R.

To specify the language in a cell, we can use magic commands, such as %python or %sql.

Notebooks also support data visualization, which allows us to create charts and tables directly from query results. This makes notebooks very useful for data exploration and analysis.

The third component is Git integration.

Databricks integrates with Git repositories, such as GitHub, GitLab, and Bitbucket. This feature is called Databricks Repos.

With Repos, developers can connect a Git repository directly to the Databricks workspace.

They can perform common Git operations such as clone, commit, pull, and push.

Using version control allows teams to track changes in code, collaborate with other developers, and maintain a stable development process.

In summary, compute clusters provide the computing power for data processing. Notebooks provide an environment for writing code and analyzing data. Git integration allows teams to manage code versions and collaborate efficiently.

Together, these components form the development environment of Databricks
